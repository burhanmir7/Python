```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
```

```python
m = len(x_train)
m
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"{x_i},{y_i}")
```

```python
plt.scatter(x_train, y_train, marker="x", c="g")
plt.title("Housing Prices")
plt.ylabel("Prices in $1000s")
plt.xlabel("Size in 1000 SqFt")
plt.show()
```

```python
w = 200
b = 100
```

```python
def calculate_model_output(w,b,x):
  m = x.shape
  f_wb = np.zeros(m)
  for i in range(len(x)):
    f_wb[i] = w * x[i] + b

  return f_wb
```

```python
tmp_f_wb = calculate_model_output(w,b,x_train)
plt.plot(x_train, tmp_f_wb, c = 'b', label="Our Prediction")
plt.scatter(x_train, y_train, marker="x", c="g", label="Actual Values")
plt.title("Housing Prices")
plt.ylabel("Prices in $1000s")
plt.xlabel("Size in 1000 SqFt")
plt.legend()
plt.show()
```
