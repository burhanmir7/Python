```python
india_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity": [80,60,78]
})

india_weather
```

```python
pakistan_weather = pd.DataFrame({
    "city" : ["karachi","lahore","peshawar"],
    "temperature":[23,48,20],
    "humidity": [56,61,39]
})

pakistan_weather
```

```python
df = pd.concat([india_weather,pakistan_weather])
df
```

```python
df = pd.concat([india_weather,pakistan_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,pakistan_weather],axis=1)
df
```

```python
temperature_df = pd.DataFrame({
    "city" : ["karachi","lahore","peshawar","multan"],
    "temperature":[23,48,20,40],
})

temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city" : ["karachi","lahore","peshawar"],
    "humidity":[67,64,34],
})

humidity_df
```

```python
df = pd.merge(humidity_df,temperature_df,on='city')
df
# gives only if all the elements are present
```

```python
df = pd.merge(humidity_df,temperature_df,on='city',how='outer')
df
```
